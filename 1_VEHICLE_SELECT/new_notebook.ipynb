```python
vehicles = ['bicycle','bike','bus', 'cars', 'cng', 'easy-bike', 'leguna', 'rickshaw', 'truck', 'van', 'Frame No.']
database = []

with open("demo","r") as f:
    l = 1
    for line in f:
        database_pre = []
        for i in range(len(vehicles)-1):
            vehicle = vehicles[i]
            start = line.find("384x640")
            if(start >= 0):
                pos = line.find(vehicle,start,)
                pos2 = line.find(vehicle+'s',start,)
                pos3 = line.find(vehicle+'es',start,)
            
                if(pos >= 0):
                    pos -= 2
                    kk = ""
                    while line[pos] != ' ':
                        kk = line[pos]+kk
                        pos -= 1
                    kk_int = int(kk)
                    if (vehicle == 'rickshaw' or vehicle == 'van' or vehicle == 'leguna' or vehicle == 'easy-bike' or vehicle == 'bicycle'):
                        database_pre.append(1)
                    else:
                        database_pre.append(0)
                elif (pos2 >= 0):
                    pos = pos2
                    pos -= 2
                    kk = ""
                    while line[pos] != ' ':
                        kk = line[pos]+kk
                        pos -= 1
                    kk_int = int(kk)
                    if (vehicle == 'rickshaw' or vehicle == 'van' or vehicle == 'leguna' or vehicle == 'easy-bike' or vehicle == 'bicycle'):
                        database_pre.append(1)
                    else:
                        database_pre.append(0)
                elif (pos3 >= 0):
                    pos = pos3
                    pos -= 2
                    kk = ""
                    while line[pos] != ' ':
                        kk = line[pos]+kk
                        pos -= 1
                    kk_int = int(kk)
                    if (vehicle == 'rickshaw' or vehicle == 'van' or vehicle == 'leguna' or vehicle == 'easy-bike' or vehicle == 'bicycle'):
                        database_pre.append(1)
                    else:
                        database_pre.append(0)
                else:
                    database_pre.append(0)
        database_pre.append(l*75)
        l += 1
        database.append(database_pre)
```
