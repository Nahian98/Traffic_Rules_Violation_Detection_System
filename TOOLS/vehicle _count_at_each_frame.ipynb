```python
vehicles = ['bicycle','bike','bus', 'cars', 'cng', 'easy-bike', 'leguna', 'rickshaw', 'track', 'van']
database = []

with open("demo","r") as f:
    for line in f:
        database_pre = []
        for vehicle in vehicles:
            start = line.find("384x640")
            if(start >= 0):
                pos = line.find(vehicle,start,)
                pos2 = line.find(vehicle+'s',start,)
                pos3 = line.find(vehicle+'es',start,)
            
                if(pos >= 0):
                    pos -= 2
                    kk = ""
                    while line[pos] != ' ':
                        kk = line[pos]+kk
                        pos -= 1
                    kk_int = int(kk)
                    database_pre.append(kk_int)
                elif (pos2 >= 0):
                    pos = pos2
                    pos -= 2
                    kk = ""
                    while line[pos] != ' ':
                        kk = line[pos]+kk
                        pos -= 1
                    kk_int = int(kk)
                    database_pre.append(kk_int)
                elif (pos3 >= 0):
                    pos = pos3
                    pos -= 2
                    kk = ""
                    while line[pos] != ' ':
                        kk = line[pos]+kk
                        pos -= 1
                    kk_int = int(kk)
                    database_pre.append(kk_int)
                else:
                    database_pre.append(0)
        database.append(database_pre)
```

```python
import pandas as pd
df = pd.DataFrame(database, columns=vehicles)
```
